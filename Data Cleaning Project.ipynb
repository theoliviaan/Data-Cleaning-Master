```python
import pandas as pd
import numpy as np
import time
import openpyxl
import xlrd
import os
import random
```

```python
# data_path = 'sales.xlsx'
# data_name = 'jan_sales'

def data_cleaning_master(data_path, data_name):

    print("Thank you for giving the details!")

    num = random.randint(1,5)

    print("Kindly hold on while we check your file path....")
    time.sleep(num)
    print("Done now")
    
    # Checking if path does not exists
    if not os.path.exists(data_path):
        print("Please enter correct path")
        return
    else: # if it exists
        # checking file type
        if data_path.endswith('.csv'):
            print("Dataset is csv!")
            data = pd.read_csv(data_path, encoding_errors='ignore')
            
        elif data_path.endswith('.xlsx'):
            print("Dataset is excel file!")
            data = pd.read_excel(data_path)
    
        else:
            print("Unknown file type")
    
    # show number or records
    # data.shape[0] is number of rows while data.shape[1] is number of columns

    num = random.randint(1,5)
    print(f"Please wait for {num} seconds, Checking total columns and rows...")
    time.sleep(num)
    print(f"Dataset contains...\nTotal rows: {data.shape[0]}\nTotal columns: {data.shape[1]}")
    
    # start cleaning
    # Find number of duplicates
    
    duplicates = data.duplicated()
    total_duplicates = data.duplicated().sum() # shows the number of duplicate rows
    # duplicated_data = data[data.duplicated()] # shows the duplicate rows
    # duplicated_data
    print(f"Dataset has {total_duplicates} total duplicate records")
    
    # saving the duplicates
    if total_duplicates > 0:
        duplicate_records = data[duplicates]
        duplicate_records.to_csv(f'{data_name}_duplicates.csv', index=None)
    
    # delete the duplicates
    df = data.drop_duplicates()
    
    # find missing values
    total_missing_values = df.isnull().sum().sum() # prints the total missing values
    missing_values_columns = df.isnull().sum()
    
    print(f"Dataset has {total_missing_values} missing values")
    
    # dealing with missing values
    columns = df.columns
    
    for col in columns:
        # print(df[col].dtype)
        if df[col].dtype in ('float64', 'int64'):# we are checking if the column is a number or float
            df[col] = df[col].fillna(df[col].mean()) # fill the space with the mean
        else:
            df.dropna(subset=col, inplace=True)
    
    # data is cleaned
    print(f"Congratulations, Dataset is cleaned! \nNumber of rows: {df.shape[0]}, Number of columns: {df.shape[1]}")
    
    # saving the clean dataset
    df.to_csv(f"{data_name}_Clean_data.csv")
    print("Dataset is saved")


if __name__ == "__main__":

    print("Welcome to Data Cleaning Master")
    # ask path and file name
    data_path = input("Please enter data path: ")
    data_name = input("Please enter dataset name: ")

    # calling the function
    data_cleaning_master(data_path, data_name)
```
